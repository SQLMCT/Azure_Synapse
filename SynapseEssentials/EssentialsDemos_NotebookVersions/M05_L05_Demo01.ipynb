## While that is running, execute the code below:

```sql
SELECT C.client_id
		,S.login_name
		,S.app_name
		,R.session_id
		,R.request_id
		,R.status
		,R.submit_time
		,R.start_time
		,R.end_compile_time
		,R.end_time
		,R.total_elapsed_time
		,R.Command
		,R.importance
		,R.resource_class
FROM sys.dm_pdw_exec_requests R
	INNER JOIN sys.dm_pdw_exec_sessions S
		ON S.session_id = R.session_id
	INNER JOIN sys.dm_pdw_exec_connections C
		ON c.session_id = S.session_id
WHERE R.status = 'running'
```

```sql
--Run this query - we expect a failure with teh group by
SELECT A.PassengerCount,
      SUM(A.TripDistanceMiles+B.TripDistanceMiles) as SumTripDistance,
      AVG(A.TripDistanceMiles+B.TripDistanceMiles) as AvgTripDistance
FROM  dbo.Trip A, dbo.Trip B 
WHERE A.TripDistanceMiles > 0 AND A.PassengerCount > 0
ORDER BY A.PassengerCount
OPTION(label='TripAvg')
```

```sql
--Investigate the failure with exec requests
Select * from sys.dm_pdw_exec_requests where status = 'failed'-- and [label] = 'TripAvg'
```
